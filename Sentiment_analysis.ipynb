```python
#remove stop words 
nltk.download("stopwords")

from nltk.corpus import stopwords
nltk.download('stopwords')
import string
from nltk.stem import PorterStemmer

sw=nltk.corpus.stopwords.words('english')

#print(sw)
```

```python
stop_words=stopwords.words('english')
punct=string.punctuation
stemmer=PorterStemmer()
```

```python
import re
cleaned_data=[]
for i in range(len(blobN)):
  #sent=re.sub('[^a-zA-Z]',' ',blob.iloc[i])
  sent=sent.lower().split()
  sent=[stemmer.stem(word) for word in sent if (word not in stop_words) and (word not in punct)]
  sent=' '.join(sent)
  cleaned_data.append(sent)
```

```python
# get the list without stop words
words_ne=[]
for word in cleaned_data:
    if word not in sw:
        words_ne.append(word)
print(words_ne)
```

```python
#Word Frequency barchat
sns.barplot()
nlp_words=nltk.FreqDist(words_ne)
nlp_words.plot(20);
```
